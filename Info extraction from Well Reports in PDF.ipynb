*Define some variables*

```python
L=len(List_key)-1
PageN=0
W=0
find=0
lineN=0
```

```python
file_Number = 0
for file in glob.glob("*.pdf"):                               # read all PDF files in working directory
    x=file.split(".")
    file_Name=x[0]
    print(file)
    print(x)
    F = open(file, 'rb')
    parser = PDFParser(F)
    document = PDFDocument(parser)
    PageNumber=resolve1(document.catalog['Pages'])['Count']       # This will give you the count of pages
    for PageN in range (0,PageNumber):
        text = extract_text(file, page_numbers = [PageN])       # open and read 1 page of the PDF file
        f = open("textfile.txt", "w",encoding="utf8",newline='')      # convert the page to a .txt file
        f.write(text)
        f.close()
        documents = []
        f = open("textfile.txt", "r",encoding="utf8",newline='')      # read the .txt file and split each single line in a list
        lines = f.read().splitlines()
        for line in lines:
            documents.append(line)
        f.close()
        LL=len(documents)
        for line in range (0,LL):
            lineN=lineN+1
            SearchMe=documents[line]
            for i in range (0,L):
                stringa=List_key[i]
                a=SearchMe.find(stringa)
                if a<0:
                    df.loc[W,i+3]=0
                else:
                    df.loc[W,i+3]="FOUND"
                    df.loc[W,L+3]=SearchMe
                    df.loc[W,0]=file_Name
                    df.loc[W,1]=PageN+1
                    df.loc[W,2]=lineN
                    find=1
            W=W+find
            find=0
        lineN=0
```
